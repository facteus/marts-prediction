```python
import pandas as pd
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2020, 5, 1)
```

```python
# Dictionary of FRED codes for downloading

fred2marts = {
    
    # Totals
    'RSAFS': 'total_sa', # Retail and Food Services, Total, Seasonally Adjusted
    'RSAFSNA': 'total_raw', # Retail and Food Services, Total
    
    'RSMVPD': 'mvpd_sa', # Motor Vehicles and Parts Dealers, Seasonally Adjusted
    'RSMVPDN': 'mvpd_raw', # Motor Vehicle and Parts Dealers
    
    'RSFSXMV': 'ex_mvpd_sa', # Retail and Food Services Excluding Motor Vehicles and Parts Dealers, Seasonally Adjusted    
    'RSFSXMVN': 'ex_mvpd_raw', # Retail and Food Services Excluding Motor Vehicles and Parts Dealers
    
    # Food Services and Drinking Places
    'RSFSDP': 'food_drink_sa', # Seasonally Adjusted
    'RSFSDPN': 'food_drink_raw',
    
    # Food and Beverage Stores
    'RSDBS': 'food_bev_sa',
    'RSDBSN': 'food_bev_raw',
    
    # Building Materials, Garden Equipment, and Supplies Dealers
    'RSBMGESD': 'build_mat_garden_sa', # Seasonally Adjusted
    'RSBMGESDN': 'build_mat_garden_raw',
    
    # Furniture and Home Furnishings Stores
    'RSFHFS': 'furniture_sa',
    'RSFHFSN': 'furniture_raw',
    
    # Nonstore Retailers
    'RSNSR': 'nonstore_sa',
    'RSNSRN': 'nonstore_raw',
    
    # General Merchandise
    'RSGMS': 'gen_merch_sa',
    'RSGMSN': 'gen_merch_raw'
}
```

```python
marts2fred = {v: k for k, v in fred2marts.items()}
```

```python
marts = web.DataReader(fred2marts.keys(), 'fred', start, end)
```

```python
marts.index = marts.index + pd.offsets.MonthEnd(0)
```

```python
marts = marts.rename(columns=fred2marts).rename_axis('date')
```
